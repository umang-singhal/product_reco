## Description


Recommend the most probable items the users can buy on their next visit to the store/website.

Control on what users have already bought, it might be the case that a single product may not be bought multiple times (do check this hypothesis)

Types of recommendations
1. Item most customers have bought
2. Complementary items which goes with what customers have bought earlier
3. <span style="color: red">Similar Items (Can explore this if we have relevant product information available)</span>

## Modeling

### Data Prep

```python
transactions = pd.read_csv("./data/reco_assignment_training.csv", parse_dates=['Tran_dt'])
holdout = pd.read_csv("./data/reco_assignment_holdout.csv", parse_dates=['Tran_dt'])
```

```python
training_users = set(transactions['Customer_num'].values.tolist())
holdout_users = set(holdout['Customer_num'].values.tolist())

holdout_extra_users = holdout_users - training_users
print(f"Extra users in holdout: {len(holdout_extra_users)}")
```

```python
training_products = set(transactions['Product_num'].values.tolist())
holdout_products = set(holdout['Product_num'].values.tolist())

holdout_extra_products = holdout_products - training_products
print(f"Extra product in holdout: {len(holdout_extra_products)}")
print("Missing Products example: ", list(holdout_extra_products)[:10])
```

We see that we have extra products in the holdout set. This could be due to two reasons - 
1. either people didn't buy any of these products in the training period 
2. or they were added during the holdout period (which is expected in ecommerce - new products are added regularly)

```python
holdout['is_missing'] = holdout['Product_num'].apply(lambda x: x in holdout_extra_products).astype('int')
holdout['is_missing'].mean()
```

```python
users_product_missing = (
    holdout
     .groupby('Customer_num')
     .agg(cnt_missing = ('is_missing', 'sum'),
          cnt_products = ('Product_num', 'count'),
         )
     .assign(atleast_1_missing = lambda x: x['cnt_missing'] > 1,
             atleast_10_pct_missing = lambda x: (x['cnt_missing'] / x['cnt_products']) > 0.1,
            )
)

print("Per Customer Purchases, where products are not in training set:")
print("atleast one: ", round(users_product_missing['atleast_1_missing'].mean() * 100, 2), "%")
print("More than 10%: ", round(users_product_missing['atleast_10_pct_missing'].mean() * 100, 2), "%")
```

As there could be multiple products which users buy from these new products 
- we can probably handle them better if we have extra information about them as how they relate to existing products like
    - category
    - description etc
    - other product related info
    
For now, we will remove them from our data and handle them separately

```python
holdout = holdout.query('is_missing == 0')

# all_users = pd.concat([transactions['Customer_num'], holdout['Customer_num']]).unique()
# all_products = pd.concat([transactions['Product_num'], holdout['Product_num']]).unique() 

all_users = transactions['Customer_num'].unique()
all_products = transactions['Product_num'].unique() 

n_users = len(all_users)
n_products = len(all_products)

u2i = dict(zip(all_users, range(n_users)))
i2u = dict(zip(range(n_users), all_users))

p2i = dict(zip(all_products, range(n_products)))
i2p = dict(zip(range(n_products), all_products))
```

```python
from scipy.sparse import coo_matrix, csr_matrix
```

```python
def generate_interactions(df, row_name: str, col_name: str, value_col: str=None):
    interaction_df = df[[row_name, col_name]].dropna().drop_duplicates()
    rows = interaction_df[row_name].values.tolist()
    cols = interaction_df[col_name].values.tolist()
    
    if value_col is None:
        data = [1] * len(rows)
    else:
        data = df[value_col]
    interactions = coo_matrix((data, (rows, cols)), shape=(n_users, n_products))
    return interactions


transactions['cust_id'] = transactions['Customer_num'].map(u2i)
transactions['product_id'] = transactions['Product_num'].map(p2i)
holdout['cust_id'] = holdout['Customer_num'].map(u2i)
holdout['product_id'] = holdout['Product_num'].map(p2i)

train = transactions.groupby(['cust_id', 'product_id'], as_index=False).agg(times_bought = ('Tran_dt', 'nunique'))
test = holdout.groupby(['cust_id', 'product_id'], as_index = False).agg(times_bought = ('Tran_dt', 'nunique'))

train_interactions = generate_interactions(train, row_name='cust_id', col_name='product_id')

# Weights = num times product bought
train_weights_interactions = generate_interactions(train, 
                                                   row_name='cust_id', 
                                                   col_name='product_id', 
                                                   value_col='times_bought')

holdout_interactions = generate_interactions(test, row_name='cust_id', col_name='product_id')
```

### Training

```python
train_interactions.shape, holdout_interactions.shape
```

```python
from lightfm import LightFM
from lightfm.evaluation import (precision_at_k, 
                                recall_at_k, 
                                reciprocal_rank,
                                auc_score
                               )
```

```python
model = LightFM(loss='warp',
                random_state=42,
                learning_rate=0.01,
                no_components=128,
                user_alpha=1e-9,
                max_sampled=5
               )

model = model.fit(train_interactions, epochs=300, verbose=False)
```

```python
model.item_embeddings.shape
```

```python
model.user_embeddings.shape
```

### Model Performance

```python
test_precision = precision_at_k(model, holdout_interactions, k=3).mean()
test_recall = recall_at_k(model, holdout_interactions, k=3).mean()
test_precision, test_recall
```

```python
test_precision = precision_at_k(model, holdout_interactions, k=10).mean()
test_recall = recall_at_k(model, holdout_interactions, k=10).mean()
test_precision, test_recall
```

```python
auc = auc_score(model, holdout_interactions)
auc.mean()
```

This shows, the items user bought were relatively predicted to be having higher probability than items which were'nt bought. <br>
For fair comparison, **AUC < .5** shows the model is poor than random selection of item for prediction. 

## Prediction

### Next 3 Products Recommendation

```python
def sample_recommendation_user(user_id,
                               model=model, 
                               interactions=train_interactions, 
                               user_dict = u2i, 
                               item_dict= i2p,
                               k = 5, 
                               show = True):
    
    n_users, n_items = interactions.shape
    user_x = user_dict[user_id]
    scores = pd.Series(model.predict(user_x, np.arange(n_items)))
    scores.index = np.arange(n_items)
    scores = list(pd.Series(scores.sort_values(ascending=False).index))
    known_items = list(np.where(interactions.tocsr()[user_x].getnnz(axis=0) > 0)[0])
    
#     scores = [x for x in scores if x not in known_items]
    known_items = [item_dict[x] for x in known_items]
    scores = [item_dict[x] for x in scores[:k]]
    
    if show == True:
        print("User: " + str(user_id))
        print("Products Bought Earlier:")
        for i, item in enumerate(known_items[:10], start=1):
            print(f"{i}. {item}")

        print("\nRecommendations:")
        for i, item in enumerate(scores, start=1):
            print(f"{i}. {item}")
            
    return scores


user_id = i2u[10]
pred = sample_recommendation_user(user_id=user_id, k=3)
```

### Similar Users

```python
def similar_users(user_id, topn=5, user_embeddings=model.user_embeddings):
    user_idx = u2i[user_id]
    user_emb = user_embeddings[user_idx]
    
    similarity = np.dot(user_embeddings, user_emb)
    top_similar_users = similarity.argsort()[::-1][: topn+1]
    similar_user_ids = [i2u[uid] for uid in top_similar_users if uid != user_idx][:topn]
#     return list(zip(similar_user_ids, similarity[top_similar_users]))
    return similar_user_ids

user = i2u[20] # or any user_id like 'C_1354499'
print(f"Similar users to: {user}")
similar_users(user, topn=10)
```

### Recall

```python
preserve_rows=False
ranks = model.predict_rank(
        holdout_interactions
    )
original_ranks = ranks.copy()

for k in [3, 5, 10]:
    ranks = original_ranks.copy()
    ranks.data = np.less(ranks.data, k, ranks.data)

    retrieved = np.squeeze(holdout_interactions.getnnz(axis=1))
    hit = np.squeeze(np.array(ranks.sum(axis=1)))

    if not preserve_rows:
        hit = hit[holdout_interactions.getnnz(axis=1) > 0]
        retrieved = retrieved[holdout_interactions.getnnz(axis=1) > 0]
    
    print(f"Recall@{k}: {(hit / retrieved).mean()*100: .2f}%")
```

We have people who buy more >10 things at a time and a few users who buys < 3 things. we can split the product recommendation objective into two objectives
 1. You may like: Predict next 3 items for users who buys small
 2. Your Usual Buys: create a bag with 7-10 items for users who buys more item in their visit and option to quick edit
     
When we have user meta and item meta, we can easily solve for cold start problems
- we can do demographics based recommendation

- product based recommendation
